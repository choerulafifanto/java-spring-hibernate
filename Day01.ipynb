# Java Spring & Hibernate

## Java

Java adalah bahasa pemrograman berorientasi objek (OOP). Java erat kaitannya dengan **classes** dan **objects**. Setiap *class* terdiri dari **attributes** (variabel) dan **methods** (function).

### 1. Class dan Object

Class merupakan sebuah "*blueprint*" untuk membuat *object*.
Contoh pembuatan class:
```java
public class MyClass {
  int x = 5;
}
```

Contoh pembuatan object:
```java
public class MyClass {
  int x = 5;

  public static void main(String[] args) {
    MyClass myObj = new MyClass();
    System.out.println(myObj.x);
  }
}
```
### 2. Static dan Public pada Method

Static = bisa diakses tanpa melakukan instansiasi (create object)
Public = bisa diakses dengan melakukan instansiasi (create object)

### 3. Java Constructor

Sebuah method untuk menginisialisasi object. Nama constructor sama dengan nama class

### 4. Java Modifier

Modifier dibagi menjadi 2:
a. Access Modifier
    - public : dapat diakses oleh class yang berbeda
    - private : hanya dapat diakses di dalam class yang sama
    - default : dapat diakses pada package yang sama
    - protected : dapat diakses pada package yang sama dan subclass
b. Non-access modifier
    - final
    - static
    - abstract
    
### 5. Java Encapsulation

Digunakan untuk menyembunyikan/menjaga data sensitive dari user. Caranya yaitu:
* Mendeklarasi atribut dengan private
* Menyediakan method **setter** dan **getter**

Contoh:
```java
public class Person {
  private String name; // private = restricted access

  // Getter
  public String getName() {
    return name;
  }

  // Setter
  public void setName(String newName) {
    this.name = newName;
  }
}
```

### 7. Java Polymorphism

Konsepnya seperti override method. Jadi satu fungsi yang sama dari beberapa class dapat melakukan banyak aksi. 
Contoh:
```java
class Animal {
  public void animalSound() {
    System.out.println("The animal makes a sound");
  }
}

class Pig extends Animal {
  public void animalSound() {
    System.out.println("The pig says: wee wee");
  }
}

class Dog extends Animal {
  public void animalSound() {
    System.out.println("The dog says: bow wow");
  }
}
```

### 8. Java Abstraction

Class untuk menyembunyikan detil tertentu dan hanya menampilkan beberapa informasi tertentu.
Abstract class dapat memiliki abstract dan regular method.

1. **Abstract class** : class yang tidak bisa digunakan untuk membuat object. Untuk mengaksesnya, harus diwarisi dari class lain.
2. **Abstract method** : method pada abstract class yang tidak memiliki body (implementasi). Body dituliskan pada subclass.

### 9. Java Interface

Interface adalah class yang seluruh method yang dimiliki tidak mempunyai body. Dengan kata lain harus diimplementasikan oleh subclass.

## Dependency Injection

Sebuah design pattern yang menerapkan prinsip IoC (Inversion of Control) yang bertujuan agar antar class tidak saling tergantung. 

![alt text](https://imgur.com/oP6I3pq.png "Logo Title Text 1")

Pada gambar di atas, semua class tidak saling terhubung langsung, melainkan dihubungkan oleh interface yang sesuai dengan karakteristik class yang terhubung. Setiap interface juga tidak saling terhubung langsung dengan class yang berbeda karakteristik, melainkan terhubung dengan interface lain yang merepresentasikan class tertentu. Hal tersebut menjadikan setiap class/interface tidak saling tergantung dengan class yang lain.
